```python
from plotnine import *
import pandas as pd

perf_nlmixr = pd.read_csv("performance_nlmixr.csv")
perf_nlmixr["algo"] = "nlmixr"
perf_pysaem = pd.read_csv("performance_pysaem.csv")
perf_pysaem["algo"] = "PySaem"

full_df = pd.concat([perf_nlmixr, perf_pysaem])
```

```python
display(full_df)
```

```python
(
    ggplot(
        full_df,
        aes(
            y="time",
            x="factor(nb_patients)",
            color="algo",
        ),
    )
    + geom_violin()
)
```
